## $\textbf{Imports}$

```python
import numpy as np
import matplotlib.pyplot as plt
```

## $\textbf{Linear convection without CFL}$

```python
def linearconv_withoutCFL(u, nx, nt, dx):
    dt = .025  
    c = 1.
    un = u
    for _ in range(nt):
        for j in range(nx):
            un[j] = u[j]
            
        for i in range(1, nx):
            u[i] = un[i] - c*(dt/dx)*(un[i] - un[i-1])
    return 0
```

## $\textbf{Linear convection with CFL}$

```python
def linearconv_withCFL(u, nx, nt, dx, CFL=0.8):
    c = 1.
    dt = CFL*dx/np.fabs(c) 
    un = np.ones(nx)
    
    for _ in range(nt):
        for j in range(nx):
            un[j] = u[j]
        for i in range(1, nx):
            u[i] = un[i] - c*(dt/dx)*(un[i] - un[i-1])
            
    return 0
```

## $\textbf{Call Funtions}$

### $\textbf{Function without CFL}$

```python
nx = 85
dx = 2 / (nx - 1)
nt = 100

grid = np.linspace(0,2,nx) 
u0 = np.ones(nx)      
u0[int(.5/dx):int(1 / dx + 1)] = 2  

u = u0.copy()
linearconv_withoutCFL(u, nx, nt, dx)

plt.plot(grid,u0)
plt.plot(grid,u)
plt.ylim([1.,2.2])
plt.xlabel('X')
plt.ylabel('Velocity')    
plt.title('1D Linear Convection') 
plt.show()
```

$\textbf{Without CFL, there is no convection.}$

### $\textbf{Function with CFL}$

```python
nx = 83
dx = 2 / (nx - 1)
nt = 40

grid = np.linspace(0,2,nx) 
u0 = np.ones(nx)      

u0[int(.5/dx):int(1 / dx + 1)] = 2  


u = u0.copy()
linearconv_withCFL(u, nx, nt, dx, CFL=0.8)

plt.plot(grid,u0)
plt.plot(grid,u)

plt.ylim([1.,2.2])
plt.xlabel('X')
plt.ylabel('Velocity')    
plt.title('1D Linear Convection') 
plt.show()
```
